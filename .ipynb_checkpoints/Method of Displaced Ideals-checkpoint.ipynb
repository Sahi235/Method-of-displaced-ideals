## Importing Libraries

```python
import numpy as np
from sklearn import preprocessing
import pandas as pd
```

## Defining our Weights, Inputs

```python
weights = np.array([6,2,4])
inputMatrix = np.array([
    [12, 12, 4854],
    [8, 3, 3442],
    [7, 4, 2776],
    [9, 2, 4270],
    [11, 8, 4450],
    [14, 6, 5830],
    [10, 8, 4667],
])
p_list = np.array([x for x in range(1, 20)])
expected_output = (inputMatrix.shape[0], p_list.shape[0])
result = np.zeros(shape=expected_output)
```

```python
def name_maker(shape, prefix):
    temp_result = np.empty(shape=shape, dtype="object")
    for i in range(shape[0]):
        temp_result[i] =  f'{prefix}{str(i+1)}'
    return temp_result
```

```python
inputMatrix
```

## Normalizing Process

```python
scalar = preprocessing.MinMaxScaler()
normalized_inp = np.round(scalar.fit_transform(inputMatrix), 2)
normalized_inp[:, 2] = 1 - normalized_inp[:, 2]
```

## Normalizing Result

```python
normalized_inp
```

```python
def l_a_calculator(arr, weights, p):
    tempArr = weights * (1 - arr)
    tempArr = tempArr ** p
    temp_arr_sum = np.sum(tempArr, axis=1)
    result = temp_arr_sum ** (1/p)
    return result
```

```python
for i in range(p_list.shape[0]):
    temp = np.round(l_a_calculator(normalized_inp, weights, p_list[i]), 2)
    result[:, i] = temp
    if (temp == result[:, i-1]).all():
        break
```

```python
column_names = np.empty(shape=result.shape[1], dtype="object")
for i in range(result.shape[1]):
    column_names[i] =  f'p={str(i+1)}'

index_names = np.empty(shape=inputMatrix.shape[0], dtype="object")
for i in range(inputMatrix.shape[0]):
    index_names[i] = f"L(A{str(i+1)})"
```

```python
final_result = pd.DataFrame(data=result, columns=column_names, index=index_names)
final_result.replace(0, np.NAN, inplace=True)
final_result.dropna(how='all', inplace=True, axis=1)
```
